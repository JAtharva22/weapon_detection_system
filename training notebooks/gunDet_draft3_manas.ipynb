```python
# FUNCTION TO FIND COMMON FILES IN FOLDERS

import os
def find_common_files(folder1, folder2):
    # Get list of file names from both folders
    files_folder1 = set(os.listdir(folder1))  # Using a set for faster comparison
    files_folder2 = set(os.listdir(folder2))

    # Find common files (file names that exist in both folders)
    common_files = files_folder1.intersection(files_folder2)

    # Check if there are common files
    if common_files:
        print(f"Common files found: {len(common_files)}")
        print("\n".join(common_files))
    else:
        print("No common files found.")
```

```python
# function to count instances of images per class
import os
from collections import defaultdict

def classes_count(labels_folder):
    class_counts = defaultdict(int)

    # Iterate through all label files
    for label_file in os.listdir(labels_folder):
        if label_file.endswith('.txt'):  # Make sure it's a label file
            label_file_path = os.path.join(labels_folder, label_file)

            # Open and read each label file
            with open(label_file_path, 'r') as f:
                lines = f.readlines()
                for line in lines:
                    # Split the line to get the class (first element in YOLO format)
                    class_id = line.split()[0]
                    class_counts[class_id] += 1

    # Print class counts
    print("Class counts in the labels folder:")
    for class_id, count in class_counts.items():
        print(f"Class {class_id}: {count} files")
```

```python
import os
import shutil

def move_all_files(src_folder, dst_folder):
    # Ensure the destination folder exists
    if not os.path.exists(dst_folder):
        os.makedirs(dst_folder)

    # Loop through all files in the source folder
    for file_name in os.listdir(src_folder):
        # Create full file paths
        src_file = os.path.join(src_folder, file_name)
        dst_file = os.path.join(dst_folder, file_name)

        # Check if it's a file and not a directory
        if os.path.isfile(src_file):
            # Move the file to the destination folder
            shutil.move(src_file, dst_file)
            print(f"Moved: {file_name}")
```

```python
# FUNCTION TO FIND MISSING FILES FROM LABELS FOLDER TO IMAGES FOLDER AND VISA VERSA

import os

def find_missing_files(labels_folder, images_folder):
    # Get list of image and label files (without extensions)
    image_files = set(f.replace('.jpg', '') for f in os.listdir(images_folder) if f.endswith('.jpg'))
    label_files = set(f.replace('.txt', '') for f in os.listdir(labels_folder) if f.endswith('.txt'))

    # Find discrepancies
    missing_labels = image_files - label_files  # Images without labels
    missing_images = label_files - image_files  # Labels without images

    # Display results
    if missing_labels:
        print(f"Images without corresponding labels: {len(missing_labels)}")
        for file in missing_labels:
            print(f"Missing label for image: {file}.jpg")

    if missing_images:
        print(f"Labels without corresponding images: {len(missing_images)}")
        for file in missing_images:
            print(f"Missing image for label: {file}.txt")

    if not missing_labels and not missing_images:
        print("All image files have corresponding label files and vice versa.")
```

```python
# FUNCTION TO FIND MATCHING FILES IN FOLDERS

import os

def find_matching_filenames(folder1, folder2):
    # Get the base filenames (without extensions) in both folders
    files_in_folder1 = {os.path.splitext(f)[0] for f in os.listdir(folder1) if os.path.isfile(os.path.join(folder1, f))}
    files_in_folder2 = {os.path.splitext(f)[0] for f in os.listdir(folder2) if os.path.isfile(os.path.join(folder2, f))}

    # Find common base filenames
    matching_files = files_in_folder1.intersection(files_in_folder2)

    if matching_files:
        print(f"Matching files found: {matching_files}")
    else:
        print("No matching files found.")

    return matching_files

# Example usage:
matching_files = find_matching_filenames('/content/10000/train', '/content/6000')
```

```python
import os

def edit_yolo_labels(folder):
    # Loop through all files in the folder
    for file_name in os.listdir(folder):
        file_path = os.path.join(folder, file_name)

        # Only process .txt files (skip non-txt files)
        if file_name.endswith(".txt") and os.path.isfile(file_path):
            with open(file_path, 'r') as file:
                lines = file.readlines()

            # Edit the first integer (the class label)
            updated_lines = []
            for line in lines:
                split_line = line.strip().split()  # Split by whitespace
                if split_line and split_line[0] == '1':  # If the first integer is '1'
                    split_line[0] = '0'  # Change it to '0'
                updated_lines.append(' '.join(split_line))  # Join the modified line back

            # Save the updated content back to the file
            with open(file_path, 'w') as file:
                file.write('\n'.join(updated_lines) + '\n')

            print(f"Updated {file_name}")

# Example usage:
edit_yolo_labels('/content/6000')
```

```python
import os
from collections import defaultdict

# Define the path to your folder (change this to match your folder structure)
folder_path = '/content/10000/train'  # Folder containing the files

# Dictionary to store filenames without extensions
file_names = defaultdict(list)

# Function to find duplicate filenames (ignoring extensions completely)
def find_duplicates(folder_path):
    for file_name in os.listdir(folder_path):
        name_without_ext = os.path.splitext(file_name)[0]  # Get the name without extension
        file_names[name_without_ext].append(file_name)

    # Find and print duplicates
    duplicates = {name: files for name, files in file_names.items() if len(files) > 1}

    if duplicates:
        print("Duplicate file names (ignoring extensions):")
        for name, files in duplicates.items():
            print(f"Filename: {name}, Files: {files}")
    else:
        print("No duplicate file names found.")

# Call the function to find duplicates
find_duplicates(folder_path)
```

```python
import os
import random
from shutil import copyfile

# Set your dataset paths
labels_folder = '/content/10000/train/labels/'  # Folder containing .txt label files
images_folder = '/content/10000/train/images/'  # Folder containing corresponding .jpg files
train_folder = '/content/train/'
valid_folder = '/content/valid/'

# Create train and valid folders if they don’t exist
os.makedirs(train_folder + 'images', exist_ok=True)
os.makedirs(train_folder + 'labels', exist_ok=True)
os.makedirs(valid_folder + 'images', exist_ok=True)
os.makedirs(valid_folder + 'labels', exist_ok=True)

# Define split ratio
split_ratio = 0.7  # 70% for training, 30% for validation

# Get list of all label files
label_files = [f for f in os.listdir(labels_folder) if f.endswith('.txt')]

# Check if label files were found
if not label_files:
    print(f"No label files found in {labels_folder}.")
else:
    print(f"Found {len(label_files)} label files.")

# Group files by class ID (the first character in the label file)
class_0_files = []
class_1_files = []

for label_file in label_files:
    file_path = os.path.join(labels_folder, label_file)
    with open(file_path, 'r') as f:
        try:
            first_line = f.readline().split()[0]  # Read the class ID from the first line
            if first_line == '0':
                class_0_files.append(label_file)
            elif first_line == '1':
                class_1_files.append(label_file)
        except IndexError:
            print(f"Skipping empty file or invalid format: {label_file}")

# Debugging: Check counts of each class
print(f"Class 0: {len(class_0_files)} files")
print(f"Class 1: {len(class_1_files)} files")

# Function to split and copy files
def split_and_copy(files, split_ratio, train_folder, valid_folder, images_folder, labels_folder):
    if not files:
        print(f"No files to process.")
        return

    random.shuffle(files)
    split_idx = int(len(files) * split_ratio)
    train_files = files[:split_idx]
    valid_files = files[split_idx:]

    # Copy to train and valid folders
    for file in train_files:
        try:
            # Copy label
            copyfile(os.path.join(labels_folder, file), os.path.join(train_folder + 'labels', file))
            # Copy image (with the same name as the label, but different extension)
            image_file = file.replace('.txt', '.jpg')
            copyfile(os.path.join(images_folder, image_file), os.path.join(train_folder + 'images', image_file))
        except FileNotFoundError:
            print(f"File not found: {file}")

    for file in valid_files:
        try:
            # Copy label
            copyfile(os.path.join(labels_folder, file), os.path.join(valid_folder + 'labels', file))
            # Copy image
            image_file = file.replace('.txt', '.jpg')
            copyfile(os.path.join(images_folder, image_file), os.path.join(valid_folder + 'images', image_file))
        except FileNotFoundError:
            print(f"File not found: {file}")

# Split and copy files for class 0 and class 1
split_and_copy(class_0_files, split_ratio, train_folder, valid_folder, images_folder, labels_folder)
split_and_copy(class_1_files, split_ratio, train_folder, valid_folder, images_folder, labels_folder)

print("Dataset has been split into training and validation sets.")
```
